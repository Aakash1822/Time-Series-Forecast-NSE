```
from nsepy import get_history
from datetime import date
import pandas as pd
```

```
from nsepy.derivatives import get_expiry_date
expire = get_expiry_date(year = 2016, month=12)
expire
```

```
tcs = get_history(symbol='TCS', start=date(2016,1,1),
                                end=date(2016,12,29))
```

```
tcs
```

```
tcs.insert(0, 'Date', pd.to_datetime(tcs.index, format='%Y-%m-%d'))
```

```
type(tcs.index)
type(tcs.Date)
tcs.Date.dt
```

```
tcs.count()
```

```
tcs.head(10)
```

```
tcs.tail(10)
```
